# Задание 3.3. Движение секущей прямой к заданной линии на плоскости (версия 1)
 


Линия на плоскости задана графиком параметрической функции $x=x(t),y=y(t)$ для $t \in [t_{min},t_{max}]$ из Задания 3.1b. Начальная точка $A$ с координатами $(x(t_{min}),y(t_{min}))$является неподвижной точкой. Точка B движется последовательно по линии от конечной точки кривой с координатами $(x(t_{max}),y(t_{max}))$ к неподвижной точке A.

$\color{red}\text{Создайте анимацию}$ движения секущей прямой, проходящей через точки А и В до момента совпадении координат точек А и В, когда секущая прямая становится касательной прямой к заданной линии в начальной точке А.

# Реализация Задания 3.3 для функции 
# $ y(x) =  sin(x)^4 $

Перед началом анимации графическая область должна содержать следующие графические объекты:

- график заданной функции, которая определяет траекторию движения
подвижной точки B;
-  неподвижную точку $A$;
-  начальное положение подвижной точки $B$;
-  cекущую прямую, проходящую через точки $A$ и$B$


```python
fig3 = plt.figure()
ax3 = plt.axes()
plt.axis([x_min-1, x_max+1, y_min-1, y_max+1])
def init():
    global point_B, secant_line 

    curve, = ax3.plot(x_array, y_array,'b')

    A = np.array([x_array[0], y_array[0]])
    point_A, = ax3.plot(A[0],A[1],'ro')

    B = np.array([x_array[-1], y_array[-1]])
    point_B, = ax3.plot(B[0],B[1],'go')

    secant_p = [A + (B - A)*t for t in [-2, 2]]
    secant_p = np.array(secant_p)
    secant_line, = ax3.plot(secant_p[:,0], secant_p[:,1],'g')

    plt.legend([r'$y(x) = \sin(x)^4$','Unmovable point A','Мovable point B','Sec'])
def at_frame3(point):
    global point_B, secant_line
    A = np.array([x_array[0], y_array[0]])
    B = np.array([point[0],point[1]])
    point_B.set_data([point[0]],[point[1]])
    if B[0]==A[0]:
        t = sympy.Symbol('t')
        y_x=sympy.sin(t)**4
        y_difx = sympy.diff(y_x, t) 
        k=y_difx.subs(t, A[0])
        y_kasstart = np.array([A[0],A[1]])
        y_kasend = np.array([A[0]+1,k+A[1]])
        y_kas=[y_kasstart+ (y_kasend-y_kasstart)*t for t in [-10, 10]]
        y_kas = np.array(y_kas)
        secant_line.set_data(y_kas[:,0], y_kas[:,1])
        secant_line.set_color('r')
    else:
        secant_p = [A + (B - A)*l for l in [-9, 9]] 
        secant_p = np.array(secant_p)
        secant_line.set_data(secant_p[:,0], secant_p[:,1])
FuncAnimation(fig3, at_frame3, frames=matrix[::-1], init_func=init, repeat=False, interval=20)   
```

Таким образом мы $\color{red}\text{написали}$ пользовательскую функцию `at_frame3(t)` , которая будет вызываться
в каждом кадре анимации, полагая, что аргумент `point` является массивом
координат подвижной точки B. $\color{red}\text{Напишем}$ документацию для `at_frame3(t)`:

Параметры:

     point : list или np.array
       Координаты точки (x, y), которые будут обновлены и визуализированы.

Глобальные переменные:

 - `point_B` : 
        Объект визуализации, представляющий точку $B$.
 - `secant_line` : 
        Объект визуализации секущей линии.

Алгоритм работы:

  1. Преобразует начальную точку A и заданную точку B в массивы numpy.
  2. Обновляет данные для точки B (`point_B.set_data`).
  3. Если точка A и точка B имеют одинаковый x-координат:
       - Рассчитывает производную заданной функции (`sin(t)^4`).
       - Вычисляет угловой коэффициент касательной.
       - Строит касательную линию через точку A.
       - Обновляет цвет касательной линии на красный.
   4. Если точки A и B имеют разные x-координаты:
       - Строит секущую линию через точки A и B.
   5. Визуализирует секущую линию с использованием `set_data`.

# Задание 3.4. Движение секущей прямой к заданной линии на плоскости (версия 2)



```python
fig4 = plt.figure()
ax4 = plt.axes()
plt.axis([x_min12-0.5, x_max12+0.5, y_min12-0.5, y_max12+0.5])
def init():
    global point_B, secant_line 
    curve, = ax4.plot(x, y,'b') 
    A = np.array([x_array[0], y_array[0]])
    point_A, = ax4.plot(A[0],A[1],'ro') 
    B = np.array([x_array[-1], y_array[-1]])
    point_B, = ax4.plot(B[0],B[1],'go') 
    secant_p = [A + (B - A)*t for t in [-0.9, 0.9]] 
    secant_p = np.array(secant_p)
    secant_line, = ax4.plot(secant_p[:,0], secant_p[:,1],'g') 
    plt.legend([r'$x = \sqrt{1-t^2}, \quad y = \tan\sqrt{1+t}$','Unmovable point A','Мovable point B','Sec'])
def at_frame3(point):
    global point_B, secant_line
    A = np.array([x_array[0], y_array[0]])
    B = np.array([point[0],point[1]])
    point_B.set_data([point[0]],[point[1]])
    if B[0]==A[0]:
        t = sympy.Symbol('t')
        x_d= sympy.sqrt(1 - t**2)
        y_d = sympy.tan(sympy.sqrt(1+t))
        x_difx=sympy.diff(x_d,t)
        y_difx = sympy.diff(y_d, t) 
        k=y_difx.subs(t, A[0])/x_difx.subs(t,A[0])
        y_kasstart = np.array([A[0],A[1]])
        y_kasend = np.array([A[0]+1,k+A[1]])
        y_kas=[y_kasstart+ (y_kasend-y_kasstart)*t for t in [-10, 10]]
        y_kas = np.array(y_kas)
        secant_line.set_data(y_kas[:,0], y_kas[:,1])
        secant_line.set_color('r')
    else:
        secant_p = [A + (B - A)*l for l in [-10, 10]] 
        secant_p = np.array(secant_p)
        secant_line.set_data(secant_p[:,0], secant_p[:,1])
FuncAnimation(fig4, at_frame3, frames=matrix12[::-1], init_func=init, repeat=False, interval=40)
```
